## School Summary
* Create an overview table that summarises key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Maths Score
  * Average Reading Score
  * % Passing Maths
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed maths **and** reading.)
  
* Create a dataframe to hold the above results

```python
school_summary = df[['school_name', 'type']].copy()
school_summary = school_summary.drop_duplicates(subset=['school_name', 'type']).reset_index(drop=True)
a = df.groupby('school_name')['Student ID'].count().reset_index()
# Rename the columns
a.columns = ['school_name', 'Total Students']
school_summary = school_summary.merge(a, on='school_name')


school_summary['Total School Budget'] = df.groupby('school_name')['budget'].mean().values
school_summary['Per Student Budget'] = school_summary['Total School Budget'] / school_summary['Total Students']
school_summary['Average Maths Score'] = df.groupby('school_name')['maths_score'].mean().values
school_summary['Average Reading Score'] = df.groupby('school_name')['reading_score'].mean().values

a = pd.DataFrame(df[df['maths_score'] >= 50].groupby('school_name')['Student ID'].count())
a.columns = ['passing math Students']
a.reset_index(inplace=True)
school_summary = school_summary.merge(a, on='school_name')
school_summary['% Passing Maths'] = (school_summary['passing math Students'] / school_summary['Total Students'])*100

a = pd.DataFrame(df[df['reading_score'] >= 50].groupby('school_name')['Student ID'].count())
a.columns = ['passing reading Students']
a.reset_index(inplace=True)
school_summary = school_summary.merge(a, on='school_name')
school_summary['% Passing Reading'] = (school_summary['passing reading Students'] / school_summary['Total Students'])*100

school_summary = school_summary.drop(['passing math Students', 'passing reading Students'], axis=1)

a = pd.DataFrame(df[passed_both].groupby('school_name')['Student ID'].count())
a.columns = ['passing Students']
a.reset_index(inplace=True)
school_summary = school_summary.merge(a, on='school_name')
school_summary['% Overall Passing'] = (school_summary['passing Students'] / school_summary['Total Students'])*100
school_summary
```

#### Top Performing Schools (By % Overall Passing)

```python
top_5_schools = school_summary.nlargest(5, '% Overall Passing').reset_index(drop=True)
top_5_schools
```
